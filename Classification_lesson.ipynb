## Feature Importance.
This is a method of scoring features to establish which features are important in developing the machine learning model. One method of scoring features is :
 ### 1. Mutual Information-
 It is the measure of the mutual dependence between two random variables. It can be implemented through scikit-learn.

```python
from sklearn.metrics import mutual_info_score
```

```python
for c in categorical:
    score = mutual_info_score(df_full_train.churn, df_full_train[c])
    print("Mutual Information for", c, "is",  score)
```

```python
# We can also do this with the apply method
def mutual_info_score_churn(series):
    return mutual_info_score(series, df_full_train.churn)
```

```python
mi = df_full_train[categorical].apply(mutual_info_score_churn)
mi.sort_values(ascending= False)
```

### 2. Correlation
While MI will be good for categorical features, correlation works for numerical features.

```python
correlation = df_full_train[numerical].corrwith(df_full_train.churn)
correlation
```
