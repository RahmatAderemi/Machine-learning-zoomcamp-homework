```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(new_df, random_state= 42, test_size= .2) 
```

```python
len(df_full_train), len(df_test)
```

```python
df_train, df_val = train_test_split(df_full_train, random_state= 42, test_size= .25)
```

```python
len(df_train), len(df_val), len(df_test)
```

```python
y_train = df_train.y.values
y_train
```

```python
del df_train["y"]
```

```python
df_train
```

```python
y_val = df_val.y.values
y_val
```

```python
del df_val["y"]
df_val
```

```python
categorical = df_train.dtypes[df_train.dtypes == "object"]
categorical= categorical.index
categorical = list(categorical)
categorical
```

```python
from sklearn.metrics import mutual_info_score
```

```python
def mutual(series):
    return mutual_info_score(series, y_train).round(2)
```

```python
df_train[categorical].apply(mutual)
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse = False)
```

```python
train_dicts = df_train[categorical + numerical].to_dict(orient = 'records')
train_dicts[1]
```

```python
dv.fit(train_dicts)
```

```python
X_train = dv.transform(train_dicts)
X_train
```

```python
dv.get_feature_names_out()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
val_dicts = df_val.to_dict(orient = "records")
val_dicts[1]
```

```python
X_val = dv.transform(val_dicts)
X_val
```

```python
y_pred = model.predict(X_val)
y_pred
```

```python
accuracy = (y_pred == y_val).mean()
accuracy
```
