```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("laptops.csv")
df
```

```python
df.isnull().sum()
```

```python
df.columns= df.columns.str.replace(' ', "_").str.lower()
df
```

```python
df1 = df[['ram','storage','screen','final_price']]
df1
```

```python
df1.isnull().sum()
```

```python
df1.screen.mean()
```

```python
df1.screen = df1.screen.fillna(15.17)
```

```python
df1.isnull().sum()
```

```python
df1.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df1, random_state= 42, test_size= 0.2)
```

```python
df_train, df_val = train_test_split(df_full_train, random_state= 42, test_size= .25)
```

```python
y_train = df_train.final_price.values
y_train
```

```python
len(df_train), len(df_val), len(df_test)
```

```python
del df_train["final_price"]
```

```python
df_train
```

```python
y_val =  df_val.final_price.values
del df_val["final_price"]
df_val
```

```python
y_test =  df_test.final_price.values
del df_test["final_price"]
df_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_unscaled = df_train.values
X_unscaled
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_unscaled)
X_train_scaled
```

```python
len(X_train_scaled)
```

```python
y_train_transformed = np.log1p(y_train)
y_train_transformed
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
```

```python
model.fit(X_train_scaled, y_train_transformed)
```

```python
X_val_unscaled = df_val.values
X_val_scaled = scaler.transform(X_val_unscaled)
X_val_scaled
```

```python
y_val_transformed = np.log1p(y_val)
```

```python
y_pred = model.predict(X_val_scaled)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error
```

```python
mae = mean_absolute_error(y_val_transformed, y_pred)
mae
```

```python
mse= mean_squared_error(y_val_transformed, y_pred)
mse
```

```python
rmse = root_mean_squared_error(y_val_transformed, y_pred)
rmse
```

```python
model.score(X_train_scaled, y_train_transformed)
```

```python
model.score(X_val_scaled, y_val_transformed)
```
